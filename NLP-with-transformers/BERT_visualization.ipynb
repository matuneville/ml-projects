```python
#!pip install bertviz
from bertviz import head_view, model_view
from transformers import BertTokenizer, BertModel, PreTrainedModel, PreTrainedTokenizer
```

```python
model = BertModel.from_pretrained('bert-base-uncased', output_attentions=True)
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased', do_lower_case = True)
```

```python
sentence_1 = "The guy who was baking the cake was Neville"
```

```python
inputs = tokenizer(sentence_1, return_tensors="pt")

outputs = model(**inputs)

attentions = outputs.attentions

token_ids = inputs['input_ids'][0] # select first and only elem of the batch

# convert tokens back to words
tokens = tokenizer.convert_ids_to_tokens(token_ids)

# visualize
head_view(attentions, tokens)
```

```python
model_view(attentions, tokens)
```
