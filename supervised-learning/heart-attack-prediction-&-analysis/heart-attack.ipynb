# Heart attack analysis and prediction

### Dataset description

- age: Age of the patient.
- sex: Sex of the patient (1 = male, 0 = female).
- cp: Chest pain type (0-3, indicating different types of chest pain).
- trtbps: Resting blood pressure (in mm Hg).
- chol: Cholesterol in mg/dl.
- fbs: Fasting blood sugar > 120 mg/dl (1 = true, 0 = false).
- restecg: Resting electrocardiographic results (0-2).
- thalachh: Maximum heart rate achieved.
- exng: Exercise-induced angina (1 = yes, 0 = no).
- oldpeak: ST depression induced by exercise relative to rest.
- slp: Slope of the peak exercise ST segment (0-2).
- caa: Number of major vessels (0-3) colored by fluoroscopy.
- thall: Thalassemia (1-3, representing different states of thalassemia).
- output: Diagnosis of heart disease (1 = presence, 0 = absence).

### Data visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
cmap = plt.get_cmap("tab10")

fig, ax = plt.subplots(nrows=2, ncols=4, figsize=(20, 10))
ax = ax.flatten()

cat_features = ['sex', 'fbs', 'exng']
ord_features = ['cp', 'thall', 'caa', 'restecg', 'slp']

# define all discrete future features just for the plotting
dis_features = cat_features + ord_features

for i in range(len(ax)):
    values = heart[dis_features[i]].value_counts()
    
    colors = cmap(np.linspace(0, 1, len(values)))
    
    wedges, texts, autotexts = ax[i].pie(values, labels=values.index,
                                         autopct='%1.1f%%', startangle=90, colors=colors)

    ax[i].set_title(dis_features[i])
    
    ax[i].legend(wedges, [f'{label} - {val} ({pct})' for label, val, pct in zip(values.index, values, [f'{pct:.1f}%' for pct in (values / values.sum() * 100)])], loc='upper right')

plt.tight_layout()
```


1. **Sex**: There are more male patients than female patients.

2. **Chest Pain Type (cp)**: The most common is type 1 (typical angina), followed by type 2 (atypical angina). Types 3 (non-anginal pain) and 4 (asymptomatic) are less frequent.

3. **Fasting Blood Sugar (fbs)**: Most patients have fasting blood sugar â‰¤ 120 mg/dl (fbs = 0). Fewer have levels > 120 mg/dl (fbs = 1).

4. **Resting Electrocardiographic Results (rest_ecg)**: Results are split between normal (0) and ST-T wave abnormalities (1). Few show left ventricular hypertrophy (2).

5. **Exercise-Induced Angina (exang)**: Most patients do not have exercise-induced angina (exang = 0), while some do (exang = 1).

6. **Slope of the Peak Exercise ST Segment (slp)**: Types 2 and 1 are majority.

7. **Number of Major Vessels (ca)**: Most patients have no major vessels colored (ca = 0). The number of patients decreases as the count of vessels increases (ca = 1, 2, or 3).

8. **Thalassemia (thall)**: Thalassemia type 2 is most common, followed by type 3. Types 1 and 0 are rare.

```python
num_features = [feat for feat in heart.columns if not feat in dis_features and feat != 'output']

colors = cmap(np.linspace(0, 1, len(num_features)))

fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(16, 10))
ax = ax.flatten()

for i in range(len(num_features)):
    sns.histplot(x=heart[num_features[i]], ax=ax[i], kde=True   , color=colors[i])
    ax[i].set_title(f'{num_features[i]} distribution')

ax[5].remove()
plt.subplots_adjust(hspace=0.4)
plt.tight_layout()
```

### Prediction

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# set target and features
y = heart['output']
X = heart[cat_features + num_features + ord_features]

# define features preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_features),
        ('cat', OneHotEncoder(), cat_features),
        ('ord', OrdinalEncoder(), ord_features)
    ])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, )

report_results = dict()
confusion_results = dict()
```

#### K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
max_accuracy, k = 0, 0

for i in range(1, 20):
    knn = Pipeline(
        steps=[
            ('preprocessor', preprocessor),
            ('knn', KNeighborsClassifier(n_neighbors=i))
        ]
    )
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    if accuracy_score(y_test, y_pred) > max_accuracy:
        k = i
        max_accuracy = accuracy_score(y_test, y_pred)
        report_results['knn'] = classification_report(y_test, y_pred)
        confusion_results['knn'] = confusion_matrix(y_test, y_pred)

print(f'k = {k}, accuracy = {max_accuracy:0.4f}')
```

#### Naive Bayes

```python
from sklearn.naive_bayes import CategoricalNB, GaussianNB
```

```python
cat_nb = Pipeline(steps=[
    ('preprocessor', ColumnTransformer(transformers=[
        ('ord', OrdinalEncoder(), ord_features),
        ('cat', OneHotEncoder(), cat_features)
    ])),
    ('classifier', CategoricalNB())
])

num_nb = Pipeline(steps=[
    ('preprocessor', ColumnTransformer(transformers=[
        ('num', StandardScaler(), num_features)
    ])),
    ('classifier', GaussianNB())
])

cat_nb.fit(X_train, y_train)
num_nb.fit(X_train, y_train)

# predict probabilities
cat_proba = cat_nb.predict_proba(X_test)
cont_proba = num_nb.predict_proba(X_test)

# combine probabilities assuming independence
combined_proba = cat_proba * cont_proba
combined_proba /= combined_proba.sum(axis=1, keepdims=True) # Normalize to make it a proper probability distribution

# get the final predictions, gets 0 or 1 depending on which column value is higher
y_pred = np.argmax(combined_proba, axis=1)

report_results['naive_bayes'] = classification_report(y_test, y_pred)
confusion_results['naive_bayes'] = confusion_matrix(y_test, y_pred)

accuracy = accuracy_score(y_test, y_pred)
print(f'accuracy: {accuracy:.4f}')
```

#### Logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = Pipeline(
        steps=[
            ('preprocessor', preprocessor),
            ('logit', LogisticRegression())
        ]
    )

lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)

report_results['logistic_regression'] = classification_report(y_test, y_pred)
confusion_results['logistic_regression'] = confusion_matrix(y_test, y_pred)

print(f'accuracy = {accuracy_score(y_test, y_pred):0.4f}')
```

### Final results

```python
fig, ax = plt.subplots(nrows=1, ncols=len(report_results), figsize=(len(report_results)*4, 3))
ax = ax.flatten()

i=0
for clf, (report, conf) in zip(report_results.keys(), zip(report_results.values(), confusion_results.values())):
    print(clf)
    print(report)
    sns.heatmap(conf, cmap='coolwarm', ax=ax[i], annot=True)
    ax[i].set_title(f'{clf} confusion matrix')
    ax[i].set_ylabel('actual outcome')
    ax[i].set_xlabel('predicted outcome')
    print('\n')
    i+=1
```

## Conclusion

KNN achieved the highest accuracy among the models tested, with a score of 87%. However, in heart disease detection, recall is more critical due to the severe consequences of false negatives. KNN also had the best recall value, with the fewest false negatives and the most true positives, resulting in a recall of 97%.
